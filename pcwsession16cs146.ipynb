```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
def get_field_goal_rate(s):
    p = sts.beta.rvs(1,1, size=s)
    return p


def field_goals(n, p):
    k = sts.binom(n, p)
    r = k.rvs(size = 1)
    print("p: " + str(p))
    print("number of field goals: " + str(r))
    plot_x = [i for i in range(n+1)]
    plot_y = k.pmf(plot_x)
    plt.bar(plot_x, plot_y)
    return k
```

```python
ps = get_field_goal_rate(3)

player1 = ps[0]
player2 = ps[1]
player3 = ps[2]
```
